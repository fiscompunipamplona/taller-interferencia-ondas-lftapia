# Ondas generadas por una piedra al caer en un estanque


```python
from pylab import plot, imshow, show
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = float(input("Digete la posicion en x en cm:"))
y1 = float(input("Digete la posicion en y en cm:"))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for xx in range( len(x) ):
    for yy in range( len(y) ):
        r = sqrt((x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]))
        outfile.write("%0.1f " % (zeta0*sin(k*r)) ) 
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("Posicion en x cm")
ylabel("Posicion en y cm")
show()
```

# Ondas generadas por dos piedras al caer en un estanque con separacion de 20cm.

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = float(input("Digete la posicion en x en cm:"))
y1 = float(input("Digete la posicion en y en cm:"))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for yy in range( len(y) ):
    for xx in range( len(x) ):
        r1 = sqrt((x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]))
        r2 = sqrt(((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]))
        A1 = zeta0*sin(k*r1)
        A2 = zeta0*sin(k*r2)
        outfile.write("%0.1f " % (A1) ) 
        outfile.write("%0.1f " % (A2) ) 
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("Posicion en x cm")
ylabel("Posicion en y cm")
show()
```
